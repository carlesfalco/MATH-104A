# Norms are continuous functions
We'll prove that a norm $\vert\vert\cdot\vert\vert$ defines a continuous function $\vert\vert\cdot\vert\vert:V\longrightarrow [0,\infty)$. For that we'll show that the norm is Lipschitz with constant 1 and from this we have the continuity. By the triangular inequality we have $$\vert\vert x\vert\vert=\vert\vert x-y+y\vert\vert\leq \vert\vert x-y\vert\vert + \vert\vert y\vert\vert\Leftrightarrow\vert\vert x\vert\vert-\vert\vert y\vert\vert\leq\vert\vert x-y\vert\vert$$
$$\vert\vert y\vert\vert=\vert\vert y-x+x\vert\vert\leq \vert\vert x-y\vert\vert + \vert\vert x\vert\vert\Leftrightarrow\vert\vert y\vert\vert-\vert\vert x\vert\vert\leq\vert\vert x-y\vert\vert$$

So we have $\big|\vert\vert x\vert\vert-\vert\vert y\vert\vert\big|\leq\vert\vert x-y\vert\vert$ proving the stated above.


# Sketching units balls for some norms at $V=\mathbb{R}^2$

If $d$ is a norm the unit ball can be defined as $B=\{x\in V\;|\;d(x)\leq 1\}$

```python
# This code produces a plot of p-norm unit ball
# INPUT: p
# OUTPUT: The expected plot
# Carles Falcó i Gandia
# 10/09/18

from random import uniform
import matplotlib.pyplot as plt
def p_norm(p,x): # x represents a point of V
    return ( abs(x[0])**p + abs(x[1])**p )**(1/p)

def p_unit_ball(p): #program for p-norm p less than infty
    unit_ballx = []
    unit_bally = []
    for i in range(0,90000):
        x1 = uniform(-1,1)
        x2 = uniform(-1,1) 
        if p_norm(p,(x1,x2)) <= 1:
            unit_ballx.append(x1)
            unit_bally.append(x2)
    plt.figure(figsize=(5,5))
    plt.plot(unit_ballx,unit_bally,'.',color = 'lightskyblue')
    plt.legend()
    plt.xlim(-1.25,1.25)
    plt.ylim(-1.25,1.25)
    plt.show()
```

So for $\vert\vert\cdot\vert\vert_2$

```python
p_unit_ball(2)
```

For $\vert\vert\cdot\vert\vert_1$

```python
p_unit_ball(1)
```

# Uniform convergence 1

Let $\{f_n\}$ be a sequence of functions on $[a,b]$ and we define $$M_n= \vert\vert f_n-f\vert\vert_{\infty}=\sup_{x\in[a,b]}|f_n(x)-f(x)|$$
We'll prove that $\{f_n\}$ converges uniformly to $f$ in $[a,b]$ if and only if $M_n$ converges to zero as $n\rightarrow\infty$. For simplicity we'll write $g_n(x)=f_n(x)-f(x)$.

We'll assume first that $\{f_n\}$ converges uniformly to $f$ in $[a,b]$ and we fix $\epsilon$. Because of the uniform convergence of the sequence of functions we can choose $N$ such that $\forall x\in[a,b]$ if $n>N$ we have $\vert g_n(x)\vert <\epsilon$. Hence we have $|M_n|=\sup_{x\in[a,b]}|g_n(x)|<\sup_{x\in[a,b]}\epsilon =\epsilon$ and we see that $M_n$ converges to zero as $n\rightarrow \infty$.

We assume now that $M_n$ converges to zero as $n\rightarrow \infty$ and we fix $\epsilon$. We take $N$ so that $M_n=\sup_{x\in[a,b]}|g_n(x)|<\epsilon$ if $n>N$. Hence, and always assuming $n>N$ we have $\vert g_n(x)\vert <\sup_{x\in[a,b]}|g_n(x)|$ by definition, and moreover $$\sup_{x\in[a,b]}|g_n(x)|<\epsilon$$ by hypothesis, what proves the stated above.

# Uniform convergence 2

(a) We'll prove that the sequence of functions $\{f_n\}$ given by: $$f_n(x)=\frac{n-1}{n}x^2+\frac{1}{n}x$$ converges uniformly to $f(x)=x^2$ on $[0,1]$. That's easy if we use what we proved in the previous section. As in it, we'll write $g_n(x)=f_n(x)-f(x)=\frac{1}{n}(x-x^2)$. Also, it is obvious that $M_n=\sup_{x\in[0,1]}|g_n(x)|=g_n(1/2)=\frac{1}{4n}$. This is because $g_n(x)\geq 0\;\forall x\in[0,1]$, $g_n(x)$ goes to $-\infty$ as $x\rightarrow\infty$ and $x\rightarrow -\infty$ and $g'_n(x)=\frac{1}{n}(1-2x)$ it is exactly zero only when $x=1/2$. Hence $g_n(1/2)$ is an absolute maximum for $g_n(x)$. Now we see that $M_n=\frac{1}{4n}\rightarrow 0$ when $n\rightarrow\infty$ so by the stated in the previous section, $\{f_n\}$ converges uniformly to $f$ on $[0,1]$.

(b) Now we ask ourselves if the sequence given by $f_n(x)=x^n$ converges uniformly on $[0,1]$. First we observe that if the sequence converges uniformly $f$ has to be $f(x)=0$ if $0\leq x< 1$ and $f(1)=1$ and we'll assume that it is uniformly convergent. Now by definition, we fix $\epsilon=0.1$ and we choose $n=N+1$ with the $N$ in the definition. We also choose $x=0.2^{1/(N+1)}<1$ and we see that:
$$\vert f_{N+1}(x)-f(x)\vert =\vert x^{N+1}\vert =0.2>0.1=\epsilon$$ which is a contradiction with the fact that the sequence was uniformly convergent. Hence $f_n(x)$ it is not uniformly convergent.

# Bernstein polynomials

of $$f(x)=\begin{cases} x,\quad 0\leq x\leq 1/2\\ 1-x,\quad 1/2<x\leq 1\end{cases}$$

```python
# Code to construct Bernstein polynomials of degree n for f
# INPUT: n degree , x variable, f function
# OUTPUT:
# Carles Falcó i Gandia
# 10/10/18

from math import factorial

# We first define the corresponding function
def f(x):
    if x >= 0 and x <= 1/2:
        return x
    elif x > 1/2 and x <= 1:
        return 1-x
    
def Bnf(n,x):
    N = int(n)
    return sum( f(k/N)*factorial(N)/factorial(k)/factorial(N-k)*x**k*(1-x)**(N-k) for k in range(0,N+1))
```

```python
# Plot of f with some bernstein polynomials
# INPUT: same as the functions before
# OUTPUT: plot 
# Carles Falcó i Gandia
# 10/10/18

from numpy import linspace, zeros
import matplotlib.pyplot as plt

colors = ['pink','y','g','b','m','c','r','orange']
plt.figure( figsize = ( 9, 4.5) )
plt.title('Plot of f and Bernstein polynomials of different degrees')
t = linspace( 0, 1, 1000)
bnft = zeros( (8, len(t)) )
ft = []
for i in t:
    ft.append(f(i))
for j in range(0,8):
    for k in range( 0, len(t) ):
        bnft[j,k] = Bnf( 2**j + 0, t[k])
    plt.plot(t, bnft[j], label = '$B_{%i}f$' %(2**j+0), color = colors[j])
plt.plot(t, ft, label = '$f$', color = 'k')
plt.legend()
plt.show()
```

We know that with this polynomials we can aproximate uniformly with arbitrary accuracy $f$. Now we'll estimate the rate of convergence of $B_nf(x)$ to $f(x)$. For computing the error we make we'll use $$\varepsilon=\vert\vert f-B_nf\vert\vert_{\infty}=\sup_{a\leq x\leq b}\vert f(x)-B_nf(x)\vert$$

```python
# We'll compute max(abs(f-bnf)) for x on [0,1]
# INPUT: t interval [0,1], nmax maximum degree of Bnf, n, diff
# OUTPUT:
# Carles Falcó i Gandia
# 10/10/18 (also the following)

t = linspace( 0, 1, 300)
nmax = 140
nd = [] #array with degrees of bnf
diff = [] #array with max(abs(f-bnf))
for i in range(1,nmax,7): #step 7 in degree
    nd.append(i)
    l = [] #all diferences for a determined degree i
    for j in t:
        l.append( abs( f(j)-Bnf(i,j) ) )
    diff.append(max(l))
```

```python
# Plot of max(abs(f-bnf)) vs n
plt.plot(nd,diff,'.')
plt.xlabel('$n$')
plt.ylabel('$\epsilon$')
plt.show()
```

Observing this we could guess a dependence for example $\max_{a\leq x\leq b}\vert f(x)-B_nf(x)\vert\propto e^{-n}$. But taking log we see the following plot

```python
# Creating plot
# Carles Falcó i Gandia
# 10/11/18
from math import log
logdiff = []
for i in diff:
    logdiff.append(log(i))
plt.plot(nd,logdiff,'.')
plt.xlabel('$n$')
plt.ylabel('$\log\epsilon$')
plt.show()
```

Which does not show a linear relation. Now assuming $\max_{a\leq x\leq b}\vert f(x)-B_nf(x)\vert\propto n^{-\alpha}$ we see the following plot:

```python
# Creating plot
# Carles Falcó i Gandia
# 10/11/18
lognd = []
for i in nd:
    lognd.append(log(i))
plt.plot(lognd,logdiff,'.')
plt.xlabel('$\log n$')
plt.ylabel('$\log \epsilon$')
plt.show()
```

Which shows a linear relation between $\log(\max_{a\leq x\leq b}\vert f(x)-B_nf(x)\vert)$ and $\log(n)$. Hence we can conclude that he Bernstein polynomials converges to $f$ at a rate of $n^{-\alpha}$ for some real positive $\alpha$. Moreover we can do a linear fit to find the value of this $\alpha$ in this case.

```python
# Linear fit between lognd and logdiff
# INPUT: the two lists
# OUTPUT: alpha
# Carles Falcó i Gandia
# 10/10/18

from scipy.stats import linregress
linregress(lognd,logdiff)
```

So we see that $\alpha\approx 0.54$. We can conclude that the error goes as $\approx \frac{1}{\sqrt{n}}$.

# Creating an integral sign $\int$

We'll use two Bézier curves to make our own integral sign. We start by choosing 5 points which will define the sign.
$$P_1 = (0,0),\;P_2 = (0,2.5),\;P_3 = (0,-2.5),\;P_4 = (1/3,3/4),\;P_5 = (-1/3,-4/3)$$
We start by plotting this points and the lines between them:
$$(x,y)=(1-t)(x_0,y_0)+t(x_1,y_1)\;\forall t\in[0,1]$$

```python
# Plot of points and lines between them
# INPUT: five points represented in two lists x_Points, y_Points
# OUTPUT: Plot
# Carles Falcó i Gandia
# 10/10/18

t = linspace(0,1,500)
x_Points = [0,0,0,1/3,-1/3]
y_Points = [0,2.5,-2.5,3/4,-3/4]
x_line1 = [(1-i)*x_Points[0]+i*x_Points[1] for i in t]
y_line1 = [(1-i)*y_Points[0]+i*y_Points[1] for i in t]
x_line2 = [(1-i)*x_Points[1]+i*x_Points[3] for i in t]
y_line2 = [(1-i)*y_Points[1]+i*y_Points[3] for i in t]
x_line3 = [(1-i)*x_Points[2]+i*x_Points[4] for i in t]
y_line3 = [(1-i)*y_Points[2]+i*y_Points[4] for i in t]
x_line4 = [(1-i)*x_Points[0]+i*x_Points[2] for i in t]
y_line4 = [(1-i)*y_Points[0]+i*y_Points[2] for i in t]
plt.figure(figsize=(5,5))
plt.plot(x_Points,y_Points,'bo',color='k')
plt.plot(x_line1,y_line1,'--',color='k')
plt.plot(x_line2,y_line2,'--',color='k')
plt.plot(x_line3,y_line3,'--',color='k')
plt.plot(x_line4,y_line4,'--',color='k')
plt.xlim(-2.8,2.8)
plt.ylim(-2.8,2.8)
plt.show()
```

Now we can compute the two Bézier quadratic curves between them by using:
$$(x,y)=(1-t)^2(x_0,y_0)+2t(1-t)(x_1,y_1)+(1-t)^2(x_2,y_2)\;\forall t\in[0,1]$$

```python
# Plot of Bézier curves, points, and tangent lines
# INPUT: The same as before
# OUTPUT: Plot
# Carles Falcó i Gandia
# 10/10/18

x_curve1 = [(1-i)**2*x_Points[0]+i*(1-i)*x_Points[1]+(1-i)**2*x_Points[3] for i in t]
y_curve1 = [(1-i)**2*y_Points[0]+i*(1-i)*y_Points[1]+(1-i)**2*y_Points[3] for i in t]
x_curve2 = [(1-i)**2*x_Points[0]+i*(1-i)*x_Points[2]+(1-i)**2*x_Points[4] for i in t]
y_curve2 = [(1-i)**2*y_Points[0]+i*(1-i)*y_Points[2]+(1-i)**2*y_Points[4] for i in t]
plt.figure(figsize=(5,5))
plt.plot(x_Points,y_Points,'bo',color='k')
plt.plot(x_line1,y_line1,'--',color='k')
plt.plot(x_line2,y_line2,'--',color='k')
plt.plot(x_line3,y_line3,'--',color='k')
plt.plot(x_line4,y_line4,'--',color='k')
plt.plot(x_curve1,y_curve1,'-',color='k')
plt.plot(x_curve2,y_curve2,'-',color='k')
plt.xlim(-2.8,2.8)
plt.ylim(-2.8,2.8)
plt.show()
```

```python
# Plot of integral sign
# Carles Falcó i Gandia
# 10/10/18

plt.figure(figsize=(5,5))
plt.plot(x_curve1,y_curve1,'-',color='k')
plt.plot(x_curve2,y_curve2,'-',color='k')
plt.xlim(-2.8,2.8)
plt.ylim(-2.8,2.8)
plt.show()
```

And we know that this curve is $C^1$ because the control point of the two Bézier curves and the point they have in common are aligned. It is $C^1$ because they share the tangent line in this common point.

$$P_2 = (0.48,2.8),\;P_3=(-0.48,2.8)$$

The mentioned points are equally aligned so the curve will be $C^1$ too.

```python
# The code and the purpose is the same as before
# we just modify the points.
# Carles Falcó i Gandia
# 10/11/18

t = linspace(0,1,500)
x_Points = [0,0.48,-0.48,1/3,-1/3]
y_Points = [0,2.8,-2.8,3/4,-3/4]
x_curve1 = [(1-i)**2*x_Points[0]+i*(1-i)*x_Points[1]+(1-i)**2*x_Points[3] for i in t]
y_curve1 = [(1-i)**2*y_Points[0]+i*(1-i)*y_Points[1]+(1-i)**2*y_Points[3] for i in t]
x_curve2 = [(1-i)**2*x_Points[0]+i*(1-i)*x_Points[2]+(1-i)**2*x_Points[4] for i in t]
y_curve2 = [(1-i)**2*y_Points[0]+i*(1-i)*y_Points[2]+(1-i)**2*y_Points[4] for i in t]
plt.figure(figsize=(5,5))
plt.plot(x_curve1,y_curve1,'-',color='k')
plt.plot(x_curve2,y_curve2,'-',color='k')
plt.xlim(-2.8,2.8)
plt.ylim(-2.8,2.8)
plt.show()
```
