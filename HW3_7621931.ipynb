# Set of best approximations to $f$ is convex

Given a linear space $V$ and $f\in V$ we'll prove that the set of best approximations to $f$ by elements in $W$ is convex. This is, given $g,h\in W$ two best approximations to $f$ the elements of $W$ $tg+(1-t)h\;\forall t\in[0,1]$ are also best approximations to $f$ by elements of $W$. If $g,h$ are best uniform approximations to $f$ and $e_n$ is the error function we know that for all $p\in W$.
$$\vert\vert f-g\vert\vert\leq\vert\vert f-p\vert\vert $$
$$\vert\vert f-h\vert\vert\leq\vert\vert f-p\vert\vert $$

Hence, by using the fact that $\vert\vert\cdot\vert\vert$ is a norm we have for each $p\in W$:
\begin{align}\vert\vert f- ( tg+(1-t)h )\vert\vert=\vert\vert tf +(1-t)f - ( tg+(1-t)h ) \vert\vert \\ 
 =\vert\vert t(f-g) +(1-t)(f-h) \vert\vert \leq t\vert\vert f-g\vert\vert +(1-t)\vert\vert f-h\vert\vert\\ 
 \leq t\vert\vert f-p\vert\vert +(1-t)\vert\vert f-p\vert\vert=\vert\vert f-p\vert\vert \end{align}
 Proving the stated above.

# Best uniform approximation to $f(x)=\sin(2x)$

by polynomials of degree at most 2 on $[0,2\pi]$.

```python
# Plot of sin(2x)
# Carles Falc√≥ i Gandia
# 15/10/18
from numpy import sin, linspace, pi
import matplotlib.pyplot as plt
x = linspace(0,2*pi,1000)
plt.plot(x,sin(2*x))
plt.show()
```

We can think about the polynomial $p(x)=0$. We observe that the error function is $\sin(2x)$ and it's obvious that the maximum error is $\sin(\pi/4)=\sin(5\pi/2)=1$. And furthermore $\sin(3\pi/4)=\sin(7\pi/4)=-1$ so this polynomial equioscillates between the maximum and the minimum of the error function $2+2$ times. By the Chebyshev Equioscillation Theorem this is a best uniform approximation to $f$ on $[0,2\pi]$ and since $p\in\mathbb{P}_2$ it is unique.

# Best uniform approximation by a constant 
to $f\in C[a,b]$.

We'll proceed as in the previous exercise. We note that a constant has degree zero so the best uniform approximation has to equioscillate between its maximum value and its minimum value at least $0+2$ times. We'll take $$p=
\frac{\max_{x\in[a,b]}f(x)+\min_{x\in[a,b]}f(x)}{2}$$
Obviously this equioscillates twice between the maximum and the minimum of the error function. Hence by the Chebysev Equioscillation Theorem $p$ is a best uniform approximation to $f$ on $[a,b]$ and since $p\in\mathbb{P}_0$ it is unique.

# Proving that a best approximation is not unique

when $V=\mathbb{R}^3$ with $\vert\vert\cdot\vert\vert_{\infty}$, $W=\text{span}\{(0,1,0),(0,0,1)\}$ and $f=(3,6,4)$.

A best approximation to $f$ in $W$ is an $(0,y,z)$ with reals $y,z$ such that $$\vert\vert f-(0,y,z)\vert\vert_{\infty}=\max\{3,\vert y-6\vert,\vert z-4\vert\}$$
so we see that for all $p\in W$ we have a lower bound $$3\leq\vert\vert f-p\vert\vert_{\infty}$$

This means that a $p$ such that $3=\vert\vert f-p\vert\vert_{\infty}$ should be a best approximation to $f$ since we can't get any other better $p$. However we have different $p$ satisfying this such as $(0,5,2)$ and $(0,6,2)$ and hence, the best approximation is not unique.

# Proving that every $p\in\mathbb{P}_n$ has a unique representation 
of the form:
$$p(x)=a_0+a_1T_1(x)+\ldots+a_nT_n(x)$$ where $T_j$ is the Chebyshev polynomial of degree $j$.

That's easy if we prove that $(1,T_1(x),\ldots,T_n(x))$ is a basis for the space $\mathbb{P}_n$. If we take the canonical basis for $\mathbb{P}_n$ and we compute the determinant of the transformation matrix between $(1,x,x^2,\ldots,x^n)$ and $(1,T_1(x),T_2(x),\ldots,T_n(x))$ we find that this is $2^{\tfrac{n(n-1)}{2}}\neq 0$. Hence we have $n+1$ linearly independent polynomials at $\mathbb{P}_n$ and that means that $(1,T_1(x),\ldots,T_n(x))$ is a basis. Also we know that the representation of a vector in a basis is unique so this proves the stated.

Alternatively we could have supposed $a_0+a_1T_1(x)+\ldots+a_nT_n(x)=b_0+b_1T_1(x)+\ldots+b_nT_n(x)$. But two polynomials are equal if and only if all of their coefficents are equal, so we have $a_i=b_i\;\forall i=0,1,\ldots,n$.

# Plotting some Chebyshev polynomials

```python
# Computation of Chebyshev polynomial of degree n
# we'll use the recursion formula: T_(n+1)=2xT_(n)-T_(n-1)
# INPUT: n degree, x variable
# OUTPUT: Chebysev polynomial of degree n
# Carles Falc√≥ i Gandia
# 16/10/18
def chebyshev_poly(n,x):
    T = [1,x]
    if n == 0:
        return 1
    k = 1
    while( k < n ):
        T.append(2*x*T[k]-T[k-1])
        k = k + 1
    return T[k]
# We can check that it works properly by doing
from sympy import var, expand
x = var('x')
for i in range(0,9):
    print('T_%i(x) =' %i,expand(chebyshev_poly(i,x)))
```

```python
# Plot of Chebyshev polynomials of degree 1,2,3 and 6
# INPUT: n degree
# OUTPUT: Plots
# Carles Falc√≥ i Gandia
# 16/10/18
from numpy import linspace
import matplotlib.pyplot as plt
X = linspace(-1.02,1.02,200)
colors = ['b','g','r']
plt.figure(figsize=(7,7))
for i in range(1,4):
    plt.plot(X,chebyshev_poly(i,X),color = colors[i-1], label = 'T_%i(x)' %i)
plt.plot(X,chebyshev_poly(6,X),color = 'orange', label = 'T_%i(x)' %6)
plt.legend(loc=0)
plt.show()
```
